```python
# Import the necessary modules
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Bring each CSV into a separate data frame
wwe_data_2013_path = '../Resources/WWE-Data-2013.csv'
wwe_data_2014_path = '../Resources/WWE-Data-2014.csv'
wwe_data_2015_path = '../Resources/WWE-Data-2015.csv'
wwe_data_2016_path = '../Resources/WWE-Data-2016.csv'
```

```python
#Read the Data
wwe_data_2013 = pd.read_csv(wwe_data_2013_path)
wwe_data_2014 = pd.read_csv(wwe_data_2014_path)
wwe_data_2015 = pd.read_csv(wwe_data_2015_path)
wwe_data_2016 = pd.read_csv(wwe_data_2016_path)
```

```python
set_2013 = set(wwe_data_2013["Wrestler"])
set_2014 = set(wwe_data_2014["Wrestler"]) 
in_one_not_other = set_2013 ^ set_2014 # the set of elements in precisely one of s1 or s2
#print(symdiff)
in_one_not_other = (in_one_not_other) - set_2013 # the set of elements in s1 but not s2
print(in_one_not_other)
```

```python
# Merge the first two datasets (2013 and 2014) on "Wrestler" so that no data is lost (should be 182 rows)
wwe_13_14_df = pd.merge(wwe_data_2013, wwe_data_2014, on='Wrestler', how='outer')
```

```python
# Rename our _x columns to "2013 Wins", "2013 Losses", and "2013 Draws"
wwe_13_14_df = wwe_13_14_df.rename(columns={'Wins_x': '2013 Wins', 'Losses_x':'2013 Losses', 'Draws_x': '2013 Draws'})
# Rename our _y columns to "2014 Wins", "2014 Losses", and "2014 Draws"
wwe_13_14_df = wwe_13_14_df.rename(columns={'Wins_y': '2014 Wins', 'Losses_y':'2014 Losses', 'Draws_y': '2014 Draws'})
wwe_13_14_df.head(2)
```
